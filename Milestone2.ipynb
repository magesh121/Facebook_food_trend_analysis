```python
#STEP 1: LOAD MILESTONE-1 DATASET
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel(r"C:\Users\HP\Downloads\infosys assignment\foodtrends_facebook_cleaned.xlsx")
df.head()
```

```python
#STEP 2: DATA OVERVIEW (EDA Step-1)
df.shape
df.info()
```

```python
#STEP 3: STATISTICAL SUMMARY (EDA Step-2)
df.describe()
#Statistical summary helps analyze central tendency and data spread.
```

```python
#STEP 4: Histogram – Engagement Distribution-Univariate
plt.hist(df['total_engagement'], bins=20)
plt.title("Total Engagement Distribution")
plt.xlabel("Total Engagement")
plt.ylabel("Frequency")
plt.show()
#This histogram shows how engagement is distributed across posts
```

```python
#STEP 5: Boxplot – Outlier Detection -Univariate
sns.boxplot(x=df['total_engagement'])
plt.title("Total Engagement Boxplot")
plt.show()
#The boxplot helps identify outliers in engagement
```

```python
#STEP 6: Engagement vs Post Type
sns.barplot(x='post_type', y='total_engagement', data=df)
plt.title("Engagement by Post Type")
plt.show()
#This shows how engagement varies between image and video posts
```

```python
#STEP 7: Engagement vs Time of Day
sns.barplot(x='time_of_day', y='total_engagement', data=df)
plt.title("Engagement by Time of Day")
plt.show()
#This helps identify the best posting time.
```

```python
#STEP 8: Correlation Heatmap-Multivariant
plt.figure(figsize=(10,6))
sns.heatmap(df[['likes_count','comments_count','shares_count','total_engagement']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
#The heatmap shows relationships between different engagement metrics.
```

```python
#Scatter Plot 1: Likes vs Total Engagement
plt.scatter(df['likes_count'], df['total_engagement'])
plt.xlabel("Likes Count")
plt.ylabel("Total Engagement")
plt.title("Likes vs Total Engagement")
plt.show()
```

```python
#Scatter Plot 2 : Comments vs Total Engagement
plt.scatter(df['comments_count'], df['total_engagement'])
plt.xlabel("Comments Count")
plt.ylabel("Total Engagement")
plt.title("Comments vs Total Engagement")
plt.show()
```

```python
#Scatter Plot 3 (Optional): Shares vs Total Engagement
plt.scatter(df['shares_count'], df['total_engagement'])
plt.xlabel("Shares Count")
plt.ylabel("Total Engagement")
plt.title("Shares vs Total Engagement")
plt.show()
```
