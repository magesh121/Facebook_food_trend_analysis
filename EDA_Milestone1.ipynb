```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('foodtrends_facebook_cleaned.xlsx')
df.head()
```

```python
df.shape
df.info()
df.describe()
```

**EDA 1: Engagement by Food Category**

```python
plt.figure(figsize=(8,5))
df.groupby('food_category')['total_engagement'].mean().sort_values().plot(kind='barh')
plt.title('Average Engagement by Food Category')
plt.xlabel('Avg Engagement')
plt.show()
```

**EDA 2: Engagement by Post Type (image vs video)**

```python
plt.figure(figsize=(5,4))
df.groupby('post_type')['total_engagement'].mean().plot(kind='bar')
plt.title('Avg Engagement by Post Type')
plt.ylabel('Avg Engagement')
plt.show()
```

**EDA 3: Engagement by Time of Day**

```python
plt.figure(figsize=(6,4))
df.groupby('time_of_day')['total_engagement'].mean().plot(kind='bar')
plt.title('Avg Engagement by Time of Day')
plt.ylabel('Avg Engagement')
plt.show()
```

**EDA 4: Engagement by Day of Week**

```python
plt.figure(figsize=(7,4))
df.groupby('day_of_week')['total_engagement'].mean().plot(kind='bar')
plt.title('Avg Engagement by Day of Week')
plt.ylabel('Avg Engagement')
plt.show()
```

**EDA 5: Reaction Distribution**

```python
reactions = ['love', 'wow', 'sad', 'angry']
df[reactions].mean().plot(kind='bar', figsize=(6,4))
plt.title('Average Reactions per Post')
plt.ylabel('Avg Count')
plt.show()
```
